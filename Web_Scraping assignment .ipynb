## 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

 Requested the source code from imdp website.

```python
data=requests.get("https://www.imdb.com/chart/top/")
data
```

```python
# converted the data in more understandable  format
soup=BeautifulSoup(data.content)
```

```python
movie_name=soup.find_all('td',class_="titleColumn")
movie_name

Name=[]
for i in movie_name:
    Name.append(i.text.replace('\n',''))
Name
    
```

```python
##GEtting the rating 

rating=soup.find_all('td',class_="ratingColumn imdbRating")
rating
```

```python
movie_rating=[]
for i in rating:
    movie_rating.append(i.text.replace('\n',''))
movie_rating

print(len(movie_rating))
```

```python
year=soup.find_all('span',class_="secondaryInfo")
year
```

```python
Year=[]
for i in year:
    Year.append(i.text.strip())
Year
```

```python
import pandas as pd
top_rated=pd.DataFrame({
    'Movie_name' : Name,
    'Rating' : movie_rating,
    'Year':  Year
})
First_100_movies=top_rated.head(100)
First_100_movies
```

# 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
from bs4 import BeautifulSoup
import requests



    
```

```python
data=requests.get('https://bookpage.com/reviews?book_genre=fiction&page=1')
data
```

```python
data.content
```

```python
soup=BeautifulSoup(data.content)
soup
```

```python
bookname=soup.find_all('h4',class_="italic")
bookname
```

```python
book_name=[]
for i in bookname:
    book_name.append(i.text.strip())
book_name
```

```python
#extracting the Author name

author=soup.find_all('p',class_="sans bold")
Author_name=[]
for i in author:
    Author_name.append(i.text.replace('\n',''))
Author_name

```

```python
#Extracting the Genre

genre=soup.find_all('p',class_="genre-links hidden-phone")
Genre=[]
for i in genre:
    Genre.append(i.text.replace('\n',''))
Genre
```

```python
#Extracting the Review
review=soup.find_all('p',class_="excerpt")
Review=[]
for i in review:
    Review.append(i.text.replace('\n',''))
Review
```

```python
#Creating the dataFrame
import pandas as pd

Book_page=pd.DataFrame({
    'book_name': book_name,
    'Author_name': Author_name,
    'Genre' :Genre,
    'Book_Review'  :Review
})
Book_page.head(5)
```

# 9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
from bs4 import BeautifulSoup
import requests
page=requests.get("https://internshala.com/fresher-jobs")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
titles=soup.find_all('div',class_="heading_4_5 profile")
# now we have all the tag we have job title.
# now we will extract the text from these tags one by one lopping over these tags.

job_titles=[]

for i in titles:
    job_titles.append(i.text.replace('\n',''))
job_titles
```

```python
title=soup.find_all('div',class_="heading_6 company_name")
title=soup.find_all('div',class_="heading_6 company_name")

Company_name=[]
for i in title:
    Company_name.append(i.text.strip())
Company_name
```

# now we are extracting the localtion

```python
location=soup.find_all('p',id="location_names")
loca_name=[]

for i in location:
    loca_name.append(i.text.strip())
loca_name
    
    
import pandas as pd
jobs=pd.DataFrame({
    'title': loca_name,
    'company': Company_name,
    'location': loca_name
})
jobs
```
